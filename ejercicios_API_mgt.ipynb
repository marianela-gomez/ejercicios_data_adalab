```python
import requests

import json

key = ""
```

```python
llamada = requests.get(f"https://api.balldontlie.io/v1/players", headers = {'Authorization': key}, params= {'next_cursor': "?cursor=NEXT_CURSOR",'per_page':100}) 

# Si la api da max 100, no le puedes pedir más (?) Pues resulta que sí, se controla con next_cursor siguiendo la doc de la API
```

```python
import pandas as pd
```

```python

list_of_players = []


for i in range(len(llamada.json()['data'])):
    list_of_players.append({"nombre" : llamada.json()['data'][i]['first_name'], "apellidos": llamada.json()['data'][i]['last_name'], "equipo": llamada.json()['data'][i]['team']['full_name']})
    
    
```

# Y ahora, poníendolo un poco más bonito

Todo el código de iterar por el json lo hago en una función llamada `get_players_info`

```python
def get_players_info(api_call):
    

    list_of_players = []

    for i in range(len(api_call.json()['data'])):
        list_of_players.append({"nombre" : api_call.json()['data'][i]['first_name'], "apellidos": api_call.json()['data'][i]['last_name'], "equipo": api_call.json()['data'][i]['team']['full_name']})
        
    return list_of_players
        
        
```

Llamo a la función con las dos opciones y lo almaceno en listas distintas

```python
list_first_100 = get_players_info(llamada)

list_second_100 = get_players_info(llamada_2)
```
